## **Theory ---**
**Data Preprocessing** - Data preprocessing refers to cleaning the for further use. While gathering the data. Our data may contain dulplicate, null, values. In data preprocessing we remove these values for easy processing. It transforms data into understandable and readable format.
*dataset.isnull()* - is used to find null rows.
*dataset.dropna()* - is used to drop the columns and rows.
But dropping is not favored because it leads to losing of data. It's is useful when there are few null values in the dataset but not recommend when their is considerable amount data to drop. Here we use IMPUTATION.

**IMPUTATION** - statistics refers to the procedure of using alternative values in place of missing data.
Done by 3 methods :-
1. Mean - replace NAN with average
2. Median - replace NAN with meadian
3. Mode - replace NAN with mode
In skew data(Unsymmetrical data) we donot use mean beacuse the mean is shifted.

**DATA STANDARDISATION** - It's used to make the dataset more accurate. For eg if we are considering a model having two features height and weight. Height is in range 1 to 2 meters and weight is in range 90 - 120 pounds. The output is biased and tilted towards weight. To minimize this type of situation we use data standardisation.
  
     HOW - Z-score
       formula :- z = (value - mean) / standard deviation
after standadisation all features have mean of 0 and standard deviation of 1.
